```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

Este código vai utilizado para fazer a analise de dados e os graficos dos dados dos projetos do PIBID de 2009 a 2022, utilizando os dados dos supervisores de cada instituição.

- Será feita a analise da distribuição da participação dos IFs em cada ano de projeto. O tratamento dos dados e a definição da intituição foi realizado préviamente no Excel


```python
total_2009_2022 = pd.read_excel('2009-2022.xlsx',sheet_name='BD_SUPERVISORES')
total_2009_2022.head()
```

```python
IFs=total_2009_2022[total_2009_2022["IF?"].str.contains("ok")] #separa apenas os dados de IFS
IFs.head()
```

Criação dos Graficos:
(como não houve participação de IFS em 2010, este ano não será representado em graficos)

```python
#Separação dos anos
ifs_2009 = IFs[IFs["ANO"] == 2009]
ifs_2011 = IFs[IFs["ANO"] == 2011]
ifs_2012 = IFs[IFs["ANO"] == 2012]
ifs_2013 = IFs[IFs["ANO"] == 2013]
ifs_2018 = IFs[IFs["ANO"] == 2018]
ifs_2020 = IFs[IFs["ANO"] == 2020]
ifs_2022 = IFs[IFs["ANO"] == 2022]
from collections import Counter
quantidade_ifs_2009=Counter(ifs_2009['SG_ENTIDADE_ENSINO'])
quantidade_ifs_2011=Counter(ifs_2011['SG_ENTIDADE_ENSINO'])
quantidade_ifs_2012=Counter(ifs_2012['SG_ENTIDADE_ENSINO'])
quantidade_ifs_2013=Counter(ifs_2013['SG_ENTIDADE_ENSINO'])
quantidade_ifs_2018=Counter(ifs_2018['SG_ENTIDADE_ENSINO'])
quantidade_ifs_2020=Counter(ifs_2020['SG_ENTIDADE_ENSINO'])
quantidade_ifs_2022=Counter(ifs_2022['SG_ENTIDADE_ENSINO'])
```

```python
#2009
fig, ax = plt.subplots(figsize=(10, 2.7))
plt.title('IFs participantes em 2009')
plt.xlabel('IFs')
plt.ylabel('Frequencia de participação em 2009')
plt.tick_params(labelsize=8 )
bar_container=plt.bar(quantidade_ifs_2009.keys(), quantidade_ifs_2009.values())
plt.bar_label(bar_container, label_type='center') 
plt.show()
```
